```python
#inserting data to db
```

```python
import csv
with open ('car.data', 'r') as f:
    car_data= csv.reader(f)
    query= 'insert into cardataset.car values ("{}","{}","{}","{}","{}","{}","{}")'
    for i in car_data:
        cursor.execute(query.format(i[0],i[1],i[2],i[3],i[4],i[5],i[6]))
mydb.commit()
        
```
